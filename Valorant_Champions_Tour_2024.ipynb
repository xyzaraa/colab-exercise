## Explanatory Data Analysis

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/VCT_2024.csv')
```

```python
df.head(7)
```

```python
df.isna().sum()
```

```python
df_clean = df.dropna()
```

```python
df_clean.isna().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df_clean['K:D'], bins=20)
plt.title('K:D Distribution')
plt.show()
```

```python
top_players = df_clean.sort_values(by='HS%', ascending=False)
top_players.reset_index(drop=True, inplace=True)
top_players.head(4)
```

```python
top_players_subset = top_players[['Player', 'HS%']]
top_players_subset.head(4)
```

Here we have top 5 players by HS average: primmie, supamen, elite, russ, medusa

```python
sns.displot(df_clean['HS%'])
plt.xlabel('Headshot Percentage')
plt.ylabel('Frekuensi')
plt.title("Head Shot")
plt.show()
```

```python
sns.lineplot(df_clean['ACS'])
plt.title("Average Combat Score")
plt.show()
```

```python
top_player_by_kill = df_clean.sort_values(by='K', ascending=False)
top_player_by_kill.reset_index(drop=True, inplace=True)
top_player_by_kill.head(10)
```

```python
top_player_by_kill_inonemap = df_clean.sort_values(by='KMax', ascending=False)
top_player_by_kill_inonemap.reset_index(drop=True, inplace=True)
top_player_by_kill_inonemap.head(5)
```

```python
sns.boxplot(x='Region', y='ACS', data=df_clean)
plt.title('Average Combat Score (ACS) by Region')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(x='Region', y='KAST', data=df_clean)
plt.title('KAST by Region')
plt.xticks(rotation=45)
plt.show()
```

```python
top_clutch_players = df_clean[['Player', 'CL%']].sort_values(by='CL%', ascending=False).head(10)
top_clutch_players
```

```python
correlation_matrix = df_clean[['ACS', 'K:D', 'KAST', 'HS%']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

## Analyzing By Team

```python
team_groups = df_clean.groupby('Team')
```

```python
team_stats = team_groups.agg({
    'K': 'mean',
    'K:D': 'mean',
    'KAST': 'mean',
    'HS%': 'mean'
}).reset_index()

team_stats.head(10)
```

```python
plt.figure(figsize=(13,7))
sns.barplot(x='Team', y='K', data=team_stats)
plt.title('Total Kill by Team')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(13,7))
sns.boxplot(data=df_clean, x='Team', y='HS%')
plt.title('Headshot Percentage by Team')
plt.xlabel('Team')
plt.ylabel('Headshot Percentage')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(13,7))
sns.boxplot(data=df_clean, x='Team', y='ACS')
plt.title('Average Combat Score Percentage by Team')
plt.xlabel('Team')
plt.ylabel('ACS Percentage')
plt.xticks(rotation=90)
plt.show()
```

```python
clutch_stats = team_groups.agg({
    'CL%': 'mean',
    'CW': 'sum',
    'CP': 'sum'
}).reset_index()

clutch_stats.head(5)
```

```python
plt.figure(figsize=(13,7))
sns.barplot(data=clutch_stats, x='Team', y='CL%')
plt.title('Avg Clutch Success Percentage per Tim')
plt.xlabel('Team Name')
plt.ylabel('Clutch Success Percentage')
plt.xticks(rotation=90)
plt.show()
```

```python
correlation_matrix = team_stats[['K', 'K:D', 'KAST']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix by Kill, KD and ACS')
plt.show()
```
